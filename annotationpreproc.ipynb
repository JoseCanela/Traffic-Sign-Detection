```
classNames = ann['Annotation tag'].unique()
classNames
```

```
classNums = {}
for i, n in enumerate(classNames):
    classNums[n] = i
```

```
from PIL import Image

def getWidthHeight(i):
    imfile = "frames/" + ann["Filename"][i]
    img = Image.open(imfile)

    w , h = img.size

    return w , h
    
def getXMid(i, w):
    x = (abs(ann["Lower right corner X"][i] + ann["Upper left corner X"][i])) / 2
    return str(x/w)

def getYMid(i, h):
    y = (abs(ann["Lower right corner Y"][i] + ann["Upper left corner Y"][i])) / 2
    return str(y/h)

def getWidth(i, w):
    return str(abs((ann["Lower right corner X"][i] - ann["Upper left corner X"][i])) / w)

def getHeight(i, w):
    return str(abs((ann["Lower right corner Y"][i] - ann["Upper left corner Y"][i])) / h)
```

```
file = open("annot.txt", 'w')

for i in range(ann.shape[0]):
    w, h = getWidthHeight(i)
    temp = str(classNums[ann["Annotation tag"][i]]) + " " + getXMid(i,w) + " " + getYMid(i,h) + " " + getWidth(i, w) + " " + getHeight(i,h) + "\n"
    file.write(temp)


```
